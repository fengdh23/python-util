### 创建

```python
# 基础
arr = np.zeros((3,4))
print(arr)
```

```python
arr = np.ones((3,4),dtype=int)
print(arr)
```

```python
arr = np.empty((3,4))
print(arr)
```

```python
arr = np.full((3,4),5) # 形状+ 赋值
print(arr)
```

```python
arr_like = np.full_like(arr,2)
print(arr_like)
```

# arange 范围

```python
arr1 = np.arange(1, 10)
arr2 = np.arange(1, 10,2) # start值 end 值 step 步长
print(arr1)
print(arr2)
```

```python
arr= np.linspace(1,10,10)
print(arr)
```

```python
arr = np.logspace(1,3,5) # 对数 间隔
print(arr)
```

# 索引和切片

```python
#  一维数组
arr = np.random.randint(1,100,20)
print(arr)
print(arr[10])
print(arr[:])
print(arr[3:7])
print(arr[slice(3,7)]) #
print(arr[arr>30])
```

```python
#  二维数组
arr = np.random.randint(1,100,(4,8)) # 4 行 8 列
print(arr)
print(arr[1,3]) # 行索引，列索引
print(arr[1,:]) # 第2行, 所有
```
